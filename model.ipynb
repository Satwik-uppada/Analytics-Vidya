```python
outside_data.rename(columns={'Title':'Course Name'},inplace=True)
outside_data.drop(columns=['Category','Page Number'],inplace=True)
```

```python
final_data = pd.merge(inside_data, outside_data,on='Course Name')
final_data
```

```python
# Function to flatten the curriculum (convert the dictionary structure to a readable string)
def flatten_curriculum(curriculum):
    try:
        # Convert the string representation of dictionary to an actual dictionary
        curriculum_dict = eval(curriculum)
        flattened_text = ""
        for chapter in curriculum_dict.values():
            flattened_text += f"Chapter: {chapter['title']}, Lessons: {', '.join(chapter['lessons'])}. "
        return flattened_text
    except Exception as e:
        return ""  # Return empty if there's an issue with the format

# Apply the flattening function to the 'Curriculum' column
final_data['Curriculum'] = final_data['Curriculum'].apply(flatten_curriculum)

# Combine Course Name, Description, and Flattened Curriculum
final_data['Combined Text'] = final_data['Course Name'] + " | " + final_data['Description'] + " |" + final_data['Curriculum']

# Display the updated DataFrame with the new column
final_data[['Course Name', 'Description', 'Curriculum', 'Combined Text']].head()
```
